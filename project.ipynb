```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['figure.dpi']=250
```

```python
hdr = pd.read_csv('project/Heart_Disease_Mortality_Data_Among_US_Adults__35___by_State_Territory_and_County.csv')
rucc = pd.read_excel('project/ruralurbancodes2013.xls')
```

```python
rucc = rucc[~(rucc['State'].isin(['AS', 'HI', 'AK', 'PR', 'MP', 'GU', 'VI']))]
```

```python
hdr = hdr[(hdr['GeographicLevel']=='County') & 
          (hdr['Stratification1']=='Overall') & 
          (hdr['Stratification2']=='Overall') & 
          (hdr['Data_Value_Footnote']!='Insufficient Data') & 
          ~(hdr['LocationAbbr'].isin(['AS', 'HI', 'AK', 'PR', 'MP', 'GU', 'VI']))]
```

```python
plt.hist(hdr['Data_Value'], bins=50);
```

```python
hdr['county_name'] = (hdr.loc[:, 'LocationDesc']
                      .str.replace('County$', '', regex=True)
                      .str.strip()
                      .str.lower())
rucc['county_name'] = (rucc.loc[:, 'County_Name']
                      .str.replace('County$', '', regex=True)
                      .str.strip()
                      .str.lower())
hdr = hdr.rename(columns={'LocationAbbr': 'state'})
rucc = rucc.rename(columns={'State': 'state'})
```

```python
df = pd.merge(rucc, hdr, on=['state', 'county_name'], how='left')
```

```python
plt.hist(df[df['RUCC_2013']==1]['Data_Value'], density=True, bins=30)
plt.hist(df[df['RUCC_2013']==9]['Data_Value'], density=True, alpha=0.5, bins=30)
```

```python
plt.plot(df.groupby('RUCC_2013').agg({'Data_Value': np.median}))
```
